```python
import pandas as pd
from datetime import datetime

# D√©finir une fonction pour calculer le nombre de mois entre deux dates
def months_between(start_date, end_date):
    # Si la date de fin n'est pas indiqu√©e, la date du jour est utilis√©e comme date de fin.
    if end_date is None:
        end_date = datetime.now()
    # Calcul de la diff√©rence de mois
    months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
    return months

# D√©finir une fonction pour g√©rer les dates de d√©but et de fin de chaque exp√©rience professionnelle et calculer la dur√©e.
def calculate_experience_duration(experience):
    start_date = datetime(experience['starts_at']['year'], experience['starts_at']['month'], experience['starts_at']['day'])
    end_date = None
    if experience['ends_at'] is not None:
        end_date = datetime(experience['ends_at']['year'], experience['ends_at']['month'], experience['ends_at']['day'])
    duration = months_between(start_date, end_date)
    return duration
```

```python
# Calculez le nombre total de mois d'exp√©rience professionnelle pour chaque ligne et ajoutez-les dans une nouvelle colonne.
df['total_experience_months(mois)'] = df['experiences'].apply(lambda exp_list: sum(calculate_experience_duration(exp) for exp in exp_list))
df
```

```python
#  D√©finir une fonction pour extraire la description
def extract_descriptions(row):
    descriptions = []
        # Extraire la description dans le champ des langues
    if isinstance(row['languages'], list):
        descriptions.extend(row['languages'])
    # Extrait la description dans le champ des comp√©tences
    if isinstance(row['skills'], list):
        descriptions.extend(row['skills'])
    # Extraire le champ r√©sum√©
    if pd.notnull(row['summary']):
        descriptions.append(row['summary'])
    # Extrait la description dans le champ exp√©riences
    if isinstance(row['experiences'], list):
        for exp in row['experiences']:
            if 'description' in exp and pd.notnull(exp['description']):
                descriptions.append(exp['description'])
    # Extrait la description dans le champ accomplishment_projects
    if isinstance(row['accomplishment_projects'], list):
        for project in row['accomplishment_projects']:
            if 'description' in project and pd.notnull(project['description']):
                descriptions.append(project['description'])
    # Renvoie une combinaison de toutes les descriptions
    return ' '.join(descriptions)

# Placer la description extraite dans une nouvelle colonne
df['all_descriptions'] = df.apply(extract_descriptions, axis=1)

print(df['all_descriptions'][6])
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
# Convertir tout le contenu de la description en minuscules
df['all_descriptions'] = df['all_descriptions'].str.lower()
# D√©finir une fonction pour nettoyer les donn√©es textuelles
def clean_text(text):
    # Supprimer la ponctuation
    text = re.sub(r'[^\w\s]', ' ', text)
    # Tokenisation
    tokens = word_tokenize(text)
    # Mots vides
    stop_words = set(stopwords.words('french'))
    # Supprimer les mots vides
    cleaned_tokens = [token for token in tokens if token not in stop_words]
    # Joindre les tokens nettoy√©s en une seule cha√Æne de caract√®res
    cleaned_text = ' '.join(cleaned_tokens)
    return cleaned_text
#nettoyer les donn√©es
df['all_descriptions_propre'] = df['all_descriptions'].apply(clean_text)
print(df['all_descriptions_propre'][0],'\n',df['all_descriptions'][0])
```

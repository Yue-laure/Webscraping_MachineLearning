```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df['DescriptionPropre'] = df['Description'].str.lower()
for i ,j in enumerate(df['DescriptionPropre']):
    # Remplacer les sauts de ligne par des pointsNL
    j=j.replace('\n','.')
    # Supprimer les caractères spéciaux à l'aide d'expressions régulières
    j = re.sub(r'[^\w\s]', ' ', j)
    # Séparer les mots
    tokens = word_tokenize(j)
    # Obtenir une liste de mots stop en français
    stop_words = set(stopwords.words('french'))
    cleaned_tokens = [token for token in tokens if token not in stop_words]
    # print(cleaned_tokens)
    # Convertir la liste de vocabulaire nettoyée en chaîne de caractères
    cleaned_text = ' '.join(cleaned_tokens)
    # Attribuer le texte nettoyé à la colonne "Description".
    # df['Description'][i]=j
    df.loc[i,'DescriptionPropre']=cleaned_text
print(df['Description'][0],'\n',df['DescriptionPropre'][0])
```

```python
df['Titre'] = df['Titre'].str.lower()
for i, titre in enumerate(df['Titre']):
    titre = re.sub(r'[^\w\s]', ' ', titre)
    tokens = word_tokenize(titre)
    stop_words = set(stopwords.words('french'))
    cleaned_tokens = [token for token in tokens if token not in stop_words]
    df.loc[i, 'TitreTraite'] = ' '.join(cleaned_tokens)
print(df['TitreTraite'],'\n',df['Titre'])
```

```python
#  Définir une fonction pour extraire la description
def extract_descriptions(row):
    descriptions = []

    # Extraire le champ Titre
    if pd.notnull(row['Titre']):
        descriptions.append(row['Titre'])
    # Extraire le champ Description
    if pd.notnull(row['Description']):
        descriptions.append(row['Description'])
    # Renvoie une combinaison de toutes les descriptions
    return ' '.join(descriptions)

# Placer la description extraite dans une nouvelle colonne
df['all_descriptions_propre'] = df.apply(extract_descriptions, axis=1)

print(df['all_descriptions_propre'][0])
```
